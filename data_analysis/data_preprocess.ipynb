```python
import pandas as pd


by_place=pd.read_csv('/home/irteam/junghye-dcloud-dir/dscapstone/data_analysis/data/5_by_place.csv')

by_place.head()
```

```python
# 첫 번째 행 제거
by_place=by_place.drop([0],axis=0,inplace=False)
by_place.head()
```

```python
by_place=by_place.drop(by_place.columns[[0]],axis=1,inplace=False)
by_place.head()
```

```python
places = list(dict.fromkeys(by_place.loc[1, '2014.1':].tolist()))
places
```

```python
transformed_data=[]

for i in range(2,len(by_place)):
    for j, year in enumerate(range(2014,2023)):
        for k,place in enumerate(places):
            year_col=f"{year}.{k}" if k>0 else str(year)
            crime=by_place.loc[i,'범죄별(2)']
            count=by_place.loc[i,year_col]
            transformed_data.append([year,place,crime,count])
            print([year,place,crime,count])
transformed_df=pd.DataFrame(transformed_data,columns=['년도', '장소', '범죄유형', '건수'])

transformed_df
```

```python
transformed_df.to_csv('/home/irteam/junghye-dcloud-dir/dscapstone/data_analysis/preprocessed_data/5_by_place.csv',index=False)
```

## 5 crimes 발생 검거

```python
crime_data=pd.read_csv('/home/irteam/junghye-dcloud-dir/dscapstone/data_analysis/data/5_crimes_b_g.csv')

crime_data.head()
```

```python
# 첫 번째 행 제거
crime_data=crime_data.drop([0],axis=0,inplace=False)
crime_data.head()
```

```python
# 첫 번째 열 제거
crime_data=crime_data.drop(crime_data.columns[[0]],axis=1,inplace=False)
```

```python
crime_data.head()
```

```python
districts=crime_data.iloc[3:,0].unique()
districts
```

```python
transformed_data=[]

for col in range(1,len(crime_data.columns),2):
    year=crime_data.columns[col].split('.')
    year=int(year[0])
    if year < 2014:
        continue
    crime_type=crime_data.loc[1,crime_data.columns[col]]
    for row in range(3,len(crime_data)):
        district=crime_data.iloc[row,0]
        occurence=crime_data.iloc[row,col]
        detection=crime_data.iloc[row,col+1]
        transformed_data.append([year,district,crime_type,occurence,detection])
        print([year,district,crime_type,occurence,detection])
```

```python
transformed_df = pd.DataFrame(transformed_data, columns=['년도', '자치구', '범죄유형', '발생건수', '검거건수'])
transformed_df
```

```python
transformed_df.to_csv('/home/irteam/junghye-dcloud-dir/dscapstone/data_analysis/preprocessed_data/5_crimes_b_g.csv',index=False)
```
